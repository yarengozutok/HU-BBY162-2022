```python
from google.colab import drive
drive.mount('/content/drive')
dosya = "/content/drive/MyDrive/Colab Notebooks/veritabanı.txt"

print ("YAREN'İN KÜTÜPHANE KATALOĞUNA HOŞGELDİNİZ")
print ("KEYİFLİ ARAMALAR DİLERİM")

class kitapkatalogu:

     def sonuc(self):
        while True:

             secim =input(" Kayıtları listelemek için: [1], \n Yeni kayıt eklemek için: [2], \n Eserin adı, yazar adı, eserin basım yılı ve yayınevine göre kayıt aramak için: [3], \n çıkıs[0] \n")

             if secim == "1":
                self.kitaplistele()
             elif secim == "2":
                self.eserekle()
             elif secim == "3":
                self.eserara()
             elif secim == "0":
                 
                break
             
            

     def __init__(self):
        books = []
        deneme = open(dosya,"r",encoding = "UTF-8")
        for i in deneme.read().strip().split("\n"):
            book = i.strip().split("|")
            books += [book]
        deneme.close()
        self.books = books
        books = None


     def eserekle(self):
        yazaradı = input("yazar adını giriniz: ")
        eseradı = input("eser adını giriniz: ")
        yayınyılı = input("Yayın yılını giriniz: ")
        yayınevı = input("Yayınevini giriniz: ")
        yenikitapekle = (yazaradı+ "|"+ eseradı + "|" + yayınevı + "|" + yayınyılı +" \n")
        print("eser bilgileriniz kataloga eklendi")
        deneme = open(dosya,"a",encoding = "UTF-8")
        deneme.write(yenikitapekle)
        deneme.close()
        

     def kitaplistele(self):
        if len(self.books) < 1:
            print("katalog da eser kaydı bulunamadı. ")
            return None

        bos  = ""
        for book in self.books:
            bos += ('eseradı:' +book[1]+ '\n')
            bos += ('yazaradı:' +book[0] +'\n')
            bos += ('yayınyılı:' +book[2]+ '\n')
            bos += ('yayınevi:' +book[3]+ '\n')
            bos += ("-"*60 + "\n")
        print(bos)

     def eserara(self):
        anahtarkelıme = input("anahtar kelime giriniz:")
        bos = ""

        for book in self.books:
            if len(self.books) > 0 and anahtarkelıme in book:
                bos += ('eseradı:' +book[1]+ '\n')
                bos += ('yazaradı:' +book[0]+ '\n')
                bos += ('yayınyılı:' +book[2]+ '\n')
                bos += ('yayınevi:' +book[3]+ '\n')
                print(bos)
        if bos== "":
            print("Maalesef katalogda aradığınız kayıt bulunamadı")
            print("İYİ GÜNLER!")

     
    
   
kitapkatalogu().sonuc()
```
